```python
import json
import os
import random
import shutil
```

```python
with open('../technology-software_template/new_html.html', 'r') as f:
    html_template = f.read()
```

```python
with open('../Placeholder_template/software_placeholders/image_placeholder.json', 'r') as f:
    image_placeholder = json.loads(f.read())
```

```python
image_placeholder
```

```python
src_root_folder = "d:\\Programing\\Work\\Electro Pi\\Customize-HTML-template-project\\Placeholder_template\\software_images"
dest_folder = "d:\\Programing\\Work\\Electro Pi\\Customize-HTML-template-project\\technology-software_template\\assets\\img\\temp"
```

```python
files = os.listdir(dest_folder)
for file in files:
  file_path = os.path.join(dest_folder, file)
  if os.path.isfile(file_path):
    os.remove(file_path)
```

```python
image_extensions={'.jpg', '.jpeg', '.png', '.bmp', '.gif'}
```

```python
for key, value in image_placeholder.items():
    k = html_template.count(key)
    
    image_folder = os.path.join(src_root_folder, value['size'])
    
    images = [f for f in os.listdir(image_folder) if os.path.splitext(f)[1].lower() in image_extensions]
    
    for _ in range(k):
        selected_image = random.choice(images)
        selected_image_path = os.path.join(image_folder, selected_image)
        
        shutil.copy2(selected_image_path, os.path.join(dest_folder, selected_image))
        
        asset_path = f'"assets/img/temp/{selected_image}"'
        print(asset_path)
        html_template = html_template.replace(key, asset_path, 1)
    
```
